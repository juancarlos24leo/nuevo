```python
#problem2
import urllib.request
import xml.etree.ElementTree as ET
import numpy as np


fhand = urllib.request.urlopen(' https://health.data.ny.gov/resource/dk4z-k3xb.xml').read()
print(type(fhand))

filexml= fhand.decode()
print(type(filexml))

tree = ET.fromstring(filexml)
#print(tree.tag)
#lst = tree.find('response')
#print(lst)

lst = tree.findall('rows/row')
print('se tiene registro ', len(lst)," medicos")

lista=[]
for item in lst:
    lista.append(int(item.find('number_of_cases').text))
npcount = np.array(lista)   
print(npcount)
print("la suma tortal es:",np.sum(npcount))
```
